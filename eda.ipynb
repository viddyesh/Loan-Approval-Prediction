```python
# store columns with specific data type
integer_columns = train_df.select_dtypes(include=['int64']).columns
float_columns = train_df.select_dtypes(include=['float64']).columns
object_columns = train_df.select_dtypes(include=['object']).columns

print("No of Integer type columns: {}".format(len(integer_columns)))
print("No of Float type columns: {}".format(len(float_columns)))
print("No of String type columns: {}".format(len(object_columns)))
```

```python
object_columns = list(object_columns)
object_columns.remove('Loan_ID')
object_columns.remove('Loan_Status')
```

```python
print(object_columns)
```

```python
#Checking if Test Data Set has any values in categorical features which is not present in Training Data Set

for object_col in object_columns:
    print(train_df[object_col].unique())
    print(test_df[object_col].unique())
```

```python
#Modifying some values in Features

train_df['Dependents'] = train_df['Dependents'].replace(['3+'],['3'])
test_df['Dependents'] = test_df['Dependents'].replace(['3+'],['3'])
train_df['Dependents'] = train_df['Dependents'].astype('int64')
test_df['Dependents'] = test_df['Dependents'].astype('int64')

train_df['Gender'] = train_df['Gender'].replace(['Male','Female'],[1,0])
test_df['Gender'] = test_df['Gender'].replace(['Male','Female'],[1,0])

train_df['Married'] = train_df['Married'].replace(['Yes','No'],[1,0])
test_df['Married'] = test_df['Married'].replace(['Yes','No'],[1,0])

train_df['Education'] = train_df['Education'].replace(['Graduate','Not Graduate'],[1,0])
test_df['Education'] = test_df['Education'].replace(['Graduate','Not Graduate'],[1,0])

train_df['Self_Employed'] = train_df['Self_Employed'].replace(['Yes','No'],[1,0])
test_df['Self_Employed'] = test_df['Self_Employed'].replace(['Yes','No'],[1,0])
```

```python
# Univariate Analysis

# Countplot of Gender
sns.countplot(x='Gender', data=train_df)
plt.title("Gender Count") #(1-Male, 0-Female)
plt.show()
```

```python
# Histogram of Loan Amt
sns.histplot(train_df['LoanAmount'], kde=True)
plt.title("Loan Amount(in Thousands) Distribution")
plt.show()
```

```python
# Histogram of Applicant Income
plt.figure(figsize=(8, 8))
plt.hist(train_df['ApplicantIncome'], bins=20)
plt.title("Histogram of Applicant Income")
plt.xlabel("Applicant Income")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar plot of Loan Amount Term
plt.figure(figsize=(5, 3))
sns.countplot(x='Loan_Amount_Term', data=train_df)
plt.title("Loan Amount Term Count")
plt.xlabel("Loan Amount Term")
plt.ylabel("Count")
plt.show()
```

```python
# Loan approval status by Property Area
sns.countplot(x='Property_Area', hue='Loan_Status', data=train_df)
plt.title("Loan Approval by Property Area")
plt.show()
```

```python
# Loan approval status by no of Dependents
sns.countplot(x='Dependents', hue='Loan_Status', data=train_df)
plt.title("Loan Approval by Dependents")
plt.show()
```

```python
# Loan approval status by Loan amt & ApplicantIncome
sns.catplot(x='ApplicantIncome', y='LoanAmount', hue='Loan_Status', data=train_df)
plt.title("Loan Amount vs. Applicant Income by Loan Approval")
plt.show()
```

```python
# Loan Approval by Gender
gender_loan_counts = train_df.groupby(['Gender', 'Loan_Status']).size().unstack()
gender_loan_counts.plot(kind='pie', subplots=True, figsize=(10, 5), autopct='%1.1f%%')
plt.title("Loan Approval by Gender")
plt.legend(['No', 'Yes'])
plt.show()
```

```python
# Loan Approval by Marital Status
marital_loan_counts = train_df.groupby(['Married', 'Loan_Status']).size().unstack()
marital_loan_counts.plot(kind='bar', stacked=True)
plt.title("Loan Approval by Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.legend(['No', 'Yes'])
plt.show()
```

```python
# Loan Approval by Education Level
edu_loan_counts = train_df.groupby(['Education', 'Loan_Status']).size().unstack()
edu_loan_counts.plot(kind='barh')
plt.title("Loan Approval by Education Level") #(1- Graduate, 0- Not Graduate)
plt.xlabel("Count")
plt.ylabel("Education Level")
plt.legend(['No', 'Yes'])
plt.show()
```

```python
# Loan distribution by Loan Amt
sns.violinplot(x='Loan_Status', y='LoanAmount', data=train_df)
plt.title("Loan Amount Distribution by Loan Approval")
plt.xlabel("Loan Approval")
plt.ylabel("Loan Amount")
plt.show()
```

```python
# Loan Approval by Credit History
credit_hist_loan_counts = train_df.groupby(['Credit_History', 'Loan_Status']).size().unstack()
credit_hist_loan_counts.plot(kind='line', stacked=True)
plt.title("Loan Approval by Credit History")
plt.xlabel("Credit History")
plt.ylabel("Count")
plt.legend(['No', 'Yes'])
plt.show()
```

```python
# Multivariate analysis

plt.figure(figsize=(10,8))
sns.heatmap(train_df.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
